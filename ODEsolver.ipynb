<h1><center>$\frac{dx}{dt}= v\times cos\Theta$</center></h1>

<h1><center>$\frac{dy}{dt}= v\times sin\Theta$</center></h1>

<h1><center>$\frac{d\Theta}{dt}= v\times \frac{tan\delta}{L}$</center></h1>

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.axisartist.axislines import SubplotZero
from matplotlib.patches import Rectangle

def car(y,t,fi):
    #constant
    l= 2.7
    v = -0.02

    #assign each ode to vector element
    ode1 = y[0]
    ode2 = y[1]
    ode3 = y[2]
    
    ode1 = v*np.cos(ode3)
    ode2 = v*np.sin(ode3)
    ode3 = ((v)*(np.tan(fi)))/l
    
    return[ode1,ode2,ode3]

#initial conditions
y0 = [7.557,3.0929,0]

#t = 13.58/0.02 = 679
#t1 = 1.44/0.02 =72 for arc
#t2 = t-t1 = 607
t = np.linspace(0,678,679)

fi = np.zeros(679)
fi[:72] = -0.5657
fi[607:] = 0.5657

# store solution
ode1 = np.empty_like(t)
ode2 = np.empty_like(t)
ode3 = np.empty_like(t)
# record initial conditions
ode1[0] = y0[0]
ode2[0] = y0[1]
ode3[0] = y0[2]

# solve ODE
for i in range(1,679):
    # span for next time step
    tspan = [t[i-1],t[i]]
    # solve for next step
    y = odeint(car,y0,tspan,args=(fi[i],))
    # store solution for plotting
    ode1[i] = y[1][0]
    ode2[i] = y[1][1]
    ode3[i] = y[1][2]
    # next initial condition
    y0 = y[1]

fig = plt.figure(figsize= (15,12))
ax = SubplotZero(fig, 111)
fig.add_subplot(ax)
ax.add_patch(Rectangle((ode1[-1]-0.89, ode2[-1]), 2.7+0.89+1.1, -1.86/2,color="yellow"))
ax.add_patch(Rectangle((ode1[-1]-0.89, ode2[-1]), 2.7+0.89+1.1, 1.86/2,color="yellow"))
ax.add_patch(Rectangle((ode1[0]-0.89, ode2[0]), 2.7+0.89+1.1, -1.86/2,color="yellow"))
ax.add_patch(Rectangle((ode1[0]-0.89, ode2[0]), 2.7+0.89+1.1, 1.86/2,color="yellow"))

# plt.xlim(-6,8)
# plt.ylim(-7,7)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.plot(ode1,ode2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Path planing using ODES");
plt.savefig('Ode for Car.png')
print(ode1[-1], ode2[-1])
```
