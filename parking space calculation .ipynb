```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_rows = 9999
df = pd.read_csv("project seminar tool final.csv") 
df
```

```python

df.plot.scatter(y='Car Length in meter', x='Min Parking Space  in meter',alpha=0.5)
```

```python
sns.lmplot(y='Car Length in meter', x='Min Parking Space  in meter',hue='Car Length in meter', data=df)
plt.title("Scatter Plot with Linear fit");
```

```python
# Linear - Line of best fit
sns.lmplot(y='Car Length in meter', x='Min Parking Space  in meter', data=df)
plt.title("Scatter Plot with Linear fit");
```

```python
#Scatter Plot with lowess line fit
sns.lmplot(y='Car Length in meter', x='Min Parking Space  in meter',lowess=True, data=df)
plt.title("Scatter Plot with Linear fit");
```

```python
sns.jointplot(x=df["Min Parking Space  in meter"], y=df["Car Length in meter"],color='red', kind='scatter')
```
